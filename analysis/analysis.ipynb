```python
import os, re
import glob

# Initialize counters
total_words = 0
total_chars = 0
session_count = 0
nltk_people_count = 0
nltk_places_count = 0
total_pages = 0

# Iterate over all text files in the output directory
for filename in glob.glob('../output/sessions/**/*.txt', recursive=True):
    session_count += 1

    with open(filename, 'r') as f:
        text = f.read()
        words = text.split()
        total_words += len(words)
        total_chars += len(text)
        pages = re.findall(r'<page>', text)
        total_pages += len(pages)

# Calculate averages
avg_words_per_session = round(total_words / session_count)
avg_chars_per_session = round(total_chars / session_count)
avg_pages_per_session = round(total_pages / session_count)

# Count people (NLTK)
with open('../output/nltk/nltk_people.txt', 'r') as f:
    for line in f:
        entity_type, _ = line.strip().split(': ')
        if entity_type.lower() == 'person':
            nltk_people_count += 1

# Count places (NLTK)
with open('../output/nltk/nltk_places.txt', 'r') as f:
    for line in f:
        entity_type, _ = line.strip().split(': ')
        if entity_type.lower() == 'place':
            nltk_places_count += 1

# Prepare the results
results = f"""
Total page count: {total_pages:,}
Total word count: {total_words:,}
Total character count: {total_chars:,}

Total number of sessions: {session_count:,}
Average page count per session: {avg_pages_per_session:,}
Average word count per session: {avg_words_per_session:,}
Average character count per session: {avg_chars_per_session:,}

Number of people mentioned (NLTK): {nltk_people_count:,}
Number of places mentioned (NLTK): {nltk_places_count:,}

"""

# Ensure the output directory exists
os.makedirs('../output/analysis', exist_ok=True)

# Write the results to a file
with open('../output/analysis/corpus_overview.txt', 'w') as f:
    f.write(results)
```
